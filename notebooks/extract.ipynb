```python
def map_col(v):
    
    v = v.replace('Director/Principal Officer','Director')
    v = v.replace('Enrollment','Enroll')
    v = v.replace('Mailing','Mail')
    v = v.replace('School Name','School')
    v = v.replace('School Classification','Religious Classification')
    v = v.replace( 'Public School District','District')
    v = v.replace('Public District','District')
    v = v.replace('Ungraded Elementary Enroll','Ungraded Elem Enroll')
    v = v.replace('Ungraded Secondary Enroll','Ungraded Sec Enroll')
    v = v.replace('Primary Email','Email')
    v = v.replace('Administrative Staff','Administrators')
    return v
  
def assign_year(n):
    *_, year = n.split('_')
    return int(year)
    
p_frames = [e.dataframe().rename(columns=map_col).assign(year=assign_year(e.name)) for e in pkg.references()]
df = pd.concat(p_frames)
df.columns
```
